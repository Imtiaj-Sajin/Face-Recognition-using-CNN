# Import Necessary Files

```python
import os
import cv2
import random
import pickle
from tqdm import tqdm
import numpy as np
import matplotlib.pyplot as plt
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
Data_Path='/content/drive/MyDrive/Imagedata'
categories= os.listdir(Data_Path)
```

**Loop through each label folder to access the images:**

```python
from PIL import Image
import pyheif

train_imgs = []
train_username = []

for c in categories:
    category_path = os.path.join(Data_Path, c)
    if len(os.listdir(category_path)) >= 20:
        for file_name in os.listdir(category_path):
            file_path = os.path.join(category_path, file_name)

            if file_name.lower().endswith(('.png', '.jpg', '.jpeg')):
                img = cv2.imread(file_path)
                if img is not None:
                    train_imgs.append(img)
                    train_username.append(c)
            elif file_name.lower().endswith('.heic'):
                heif_file = pyheif.read(file_path)
                image = Image.frombytes(
                    heif_file.mode,
                    heif_file.size,
                    heif_file.data,
                    "raw",
                    heif_file.mode,
                    heif_file.stride,
                )
                img = np.array(image)
                train_imgs.append(img)
                train_username.append(c)
            else:
                continue
```
